```python
#Cleaning and wrangling of data 

rows=df.shape[0]
columns=df.shape[1]
print("Number of rows=", rows)
print("Number of columns=", columns)
df.replace("NaN", pd.NA, inplace=True)
print(df.columns.tolist())
missing=df.isnull().sum()
print("")
print("Total missing values in each column in the given dataset")
print(missing)
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.drop_duplicates()
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]
print(df.info())
```

```python
#Visualisation of the dataset

print(df.describe())

#Two main libraries, matplotlib and seaborn is used for the visualization in order to use graphs and charts

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 4))
plt.plot(df['date'], df['close'])
plt.title('Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
sns.histplot(df['volume'], bins=50, kde=True)
plt.title('Volume Distribution')
plt.show()

sns.boxplot(x=df['close'])
plt.title('Close Price Outliers')
plt.show()


plt.figure(figsize=(8, 6))
sns.heatmap(df[['open', 'high', 'low', 'close', 'volume']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('AAPL_5yr_stock_data.csv')
df['date'] = pd.to_datetime(df['date'], errors='coerce')
numeric_columns = ['open', 'close']
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df = df.dropna(subset=['open', 'close'])
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='open', y='close', alpha=0.6, color='dodgerblue')
plt.title('Scatter Plot: Open vs Close Price')
plt.xlabel('Open Price')
plt.ylabel('Close Price')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('AAPL_5yr_stock_data.csv')


df['close'] = pd.to_numeric(df['close'], errors='coerce')
df['volume'] = pd.to_numeric(df['volume'], errors='coerce')

df = df.dropna(subset=['close', 'volume'])

plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x=df['close'], color='skyblue')
plt.title('Boxplot of Close Prices')
plt.xlabel('Close Price')
plt.subplot(1, 2, 2)
sns.boxplot(x=df['volume'], color='salmon')
plt.title('Boxplot of Trading Volume')
plt.xlabel('Volume')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('AAPL_5yr_stock_data.csv')
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['day_of_week'] = df['date'].dt.day_name()
df['volume'] = pd.to_numeric(df['volume'], errors='coerce')
df = df.dropna(subset=['day_of_week', 'volume'])
plt.figure(figsize=(10, 6))
sns.boxplot(x='day_of_week', y='volume', data=df, palette='Set3')
plt.title('Boxplot of Trading Volume by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Trading Volume')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

df['year'] = df['date'].dt.year
plt.figure(figsize=(10, 6))
sns.boxplot(x='year', y='close', data=df, palette='Blues')
plt.title('Boxplot of Close Prices by Year')
plt.xlabel('Year')
plt.ylabel('Close Price')
plt.tight_layout()
plt.show()
```

```python
df['quarter'] = df['date'].dt.to_period('Q').astype(str)
plt.figure(figsize=(12, 6))
sns.boxplot(x='quarter', y='close', data=df, palette='PuBuGn')
plt.title('Boxplot of Close Prices by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
